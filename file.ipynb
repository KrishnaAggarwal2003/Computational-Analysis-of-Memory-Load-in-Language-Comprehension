```python
import networkx as nx
import matplotlib.pyplot as plt
import spacy

sentence = "John threw the trash can into the bin"
nlp_eng = spacy.load("en_core_web_sm")
doc = nlp_eng(sentence)

# Define the edges of the graph (tuples of connected nodes)
edges = [(f"{token.head.text}_{token.head.i}", f"{token.text}_{token.i}") for token in doc]
edge_labels = {(f"{token.text}_{token.i}", f"{token.head.text}_{token.head.i}"): f'{token.dep_}' for token in doc}

# Create the graph
G = nx.DiGraph()
G.add_edges_from(edges)

# Define node labels to show only the text (without indices)
node_labels = {f"{token.text}_{token.i}": token.text for token in doc}

# Generate positions for the nodes in the graph
pos = nx.spring_layout(G)

# Draw the graph
plt.figure(figsize=(8,8))
nx.draw(G, pos, labels=node_labels, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Generate the adjacency matrix
nodelist = [f"{token.text}_{token.i}" for token in doc]
adj_matrix = nx.to_numpy_array(G, nodelist=nodelist)

# Display the adjacency matrix
print("Adjacency Matrix:\n", adj_matrix, '\n')
print(f'Shape of matrix = {adj_matrix.shape}\n')

# Show the plot
plt.show()
```

```python
import tarfile
import os

# Path to your .tgz file
tgz_path = "deep-ud-2.8-data.tgz"

# Output directory for extracted files
output_dir = "extracted_treebank"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Open and extract only .conllu files
with tarfile.open(tgz_path, "r:gz") as tar:
    for member in tar.getmembers():
        if member.name.endswith(".conllu"):
            tar.extract(member, path=output_dir)
            print(f"Extracted: {member.name}")
```
